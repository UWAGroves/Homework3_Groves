```python
# Link Google Drive

from google.colab import drive
drive.mount('/content/drive')

filedir = "/content/drive/MyDrive/"

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Import data

fname = filedir + "MaunaLoaCO2_monthly_sitename.csv"
df = pd.read_csv(fname)

df

```

```python
#Remove years 1958-2022
df_filtered = df[~df['Year'].isin([1958, 2022])]

print(df_filtered)
```

```python
# Remove rows where "numofdays" or "stdofdays" are negative

df_filtered = df_filtered[(df_filtered['numofdays'] >= 0) & (df['stdofdays'] >= 0)]

print(df_filtered)
```

```python
# Calculate annual average CO₂ concentration
annual_average = df_filtered.groupby('Year', as_index=False)['Monthly_average'].mean()

# Rename column
annual_average.rename(columns={'Monthly_average': 'Annual_average'}, inplace=True)

print(annual_average)
```

```python
# Calculate annual average deseasonalized CO₂ concentration
deseasonalized_annual_average = df_filtered.groupby('Year', as_index=False)['deseasonalized'].mean()

# Rename column
deseasonalized_annual_average.rename(columns={'deseasonalized': 'Annual_average'}, inplace=True)

print(deseasonalized_annual_average)
```

```python
# Compute the standard deviation for each year's Monthly_average
std_of_months = df_filtered.groupby('Year', as_index=False)['Monthly_average'].std()

# Rename column
std_of_months.rename(columns={'Monthly_average': 'std_of_months'}, inplace=True)

print(std_of_months)
```

```python
# Function to classify CO₂ levels
def CatC02 (i):
    if i < 350:
        return "Low"
    elif 350 <= i <= 400:
        return "Moderate"
    else:
        return "High"

# Apply function to create new column
df_filtered["CO2_Category"] = df_filtered["Monthly_average"].apply(CatC02)
print(df_filtered)
```

```python
# Save df_filtered as a CSV file
df_filtered.to_csv("Groves_MaunaLoaCO2_annual_Hw3.csv", index=False)
```
